```python
#!/usr/bin/env python2
# -*- coding: utf-8 -*-
"""
Created on Sat Apr 29 16:04:07 2017

@author: peterweber
"""

import pandas as pd
import numpy as np
from collections import Counter
from nltk.corpus import stopwords
```

```python
df_train = pd.read_csv("~/Projects/Kaggle/Kaggle_QuoraQuestionPairs/data/train.csv").iloc[0:100]
df_test  = pd.read_csv("~/Projects/Kaggle/Kaggle_QuoraQuestionPairs/data/test.csv").iloc[0:100]
```

```python
def get_weight(count, eps=10, min_count=2):
    return 0 if count < min_count else 1 / float(count + eps)
```

```python
train_qs = pd.Series(df_train['question1'].tolist() + df_train['question2'].tolist()).astype(str)
words = (" ".join(train_qs)).lower().split()
counts = Counter(words)
weights = {word: get_weight(count) for word, count in counts.items()}

#stops = set(stopwords.words("english"))
```

```python
def word_shares(row):
    q1 = set(str(row['question1']).lower().split())
    #q1words = q1.difference(stops)
    #if len(q1words) == 0:
    #return '0:0:0:0:0'

    q2 = set(str(row['question2']).lower().split())
    #q2words = q2.difference(stops)
    #if len(q2words) == 0:
    #return '0:0:0:0:0'

    #q1stops = q1.intersection(stops)
    #q2stops = q2.intersection(stops)

    shared_words = q1.intersection(q2)
    shared_weights = [weights.get(w, 0) for w in shared_words]
    total_weights = [weights.get(w, 0) for w in q1] + [weights.get(w, 0) for w in q2]

    R1 = np.sum(shared_weights) / np.sum(total_weights) #tfidf share
    R2 = len(shared_words) / (len(q1) + len(q2)) #count share
    R31 = len(q1) / len(q1) #stops in q1
    R32 = len(q2) / len(q2) #stops in q2
    return '{}:{}:{}:{}:{}'.format(R1, R2, len(shared_words), R31, R32)
```

```python
df = pd.concat([df_train, df_test])
df['word_shares'] = df.apply(word_shares, axis=1, raw=True)

x = pd.DataFrame()

x['word_match'] = df['word_shares'].apply(lambda x: float(x.split(':')[0]))
x['tfidf_word_match'] = df['word_shares'].apply(lambda x: float(x.split(':')[1]))
x['shared_count'] = df['word_shares'].apply(lambda x: float(x.split(':')[2]))
x['stops1_ratio'] = df['word_shares'].apply(lambda x: float(x.split(':')[3]))
x['stops2_ratio'] = df['word_shares'].apply(lambda x: float(x.split(':')[4]))
```
